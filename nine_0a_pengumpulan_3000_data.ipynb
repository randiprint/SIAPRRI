# Scrapping Data 3000 butir

```python
%pip install google_play_scraper
# Perintah untuk menginstall library `google_play_scraper` langsung dari Jupyter Notebook.
# Kalau dijalankan di script Python biasa, cukup gunakan `pip install google_play_scraper` di terminal.

from google_play_scraper import app, Sort, reviews
# Mengimpor fungsi-fungsi utama dari library `google_play_scraper`.
# - app     : untuk mengambil detail aplikasi (nama, rating, dll).
# - Sort    : untuk mengatur urutan review (misalnya terbaru atau paling relevan).
# - reviews : untuk mengambil ulasan pengguna.

import pandas as pd
# Mengimpor pandas untuk pengolahan data tabular (dataframe).

import numpy as np
# Mengimpor numpy untuk manipulasi data numerik (array, vektor, dll).

app_id = 'com.Info_BMKG'
# Menentukan ID aplikasi di Google Play Store.
# ‚ö†Ô∏è Catatan: pastikan app_id benar sesuai package name di Play Store.
# Untuk aplikasi Info BMKG biasanya `id.go.bmkg.infoBMKG`.

result, continuation_token = reviews(
    app_id,
    lang='id',           # Bahasa ulasan yang diambil. Default: 'en'. Di sini pakai bahasa Indonesia.
    country='id',        # Negara asal review. Default: 'us'. Di sini diatur Indonesia.
    sort=Sort.NEWEST,    # Mengurutkan review berdasarkan yang terbaru.
    count=3000,          # Jumlah ulasan yang ingin diambil. Default 100, di sini minta 3000.
    filter_score_with=None,  # Filter berdasarkan rating.
                             # None = semua bintang (1-5).
                             # Bisa isi 1 untuk review bintang 1 saja, dst.
)
# Fungsi ini mengembalikan dua output:
# - result              : list of dict berisi data ulasan.
# - continuation_token  : token untuk melanjutkan ambil data berikutnya (pagination).

df_mentah = pd.DataFrame(np.array(result), columns=['review'])
# Mengubah hasil list of dict menjadi DataFrame pandas.
# Data awal dimasukkan ke kolom tunggal bernama 'review'.

df_mentah = df_mentah.join(pd.DataFrame(df_mentah.pop('review').tolist()))
# Memecah kolom 'review' yang masih berupa dictionary ke dalam kolom-kolom terpisah
# (seperti 'content', 'score', 'userName', 'at', dll).

df_mentah.to_csv('sample_data/the_data_selected.csv', index=False)
# Menyimpan DataFrame hasil ke dalam file CSV bernama 'the_data_selected.csv'
# di folder 'sample_data' tanpa menyertakan index.
```

# Filter dengan waktu

```python
from datetime import datetime

# Use the correct DataFrame
df = df_mentah

# Filter review from 1 January 2024 to 30 January 2025
start_date = datetime(2024, 1, 1)
end_date = datetime(2025, 1, 30)
df_filtered = df[(df['at'] >= start_date) & (df['at'] <= end_date)]

# Save to CSV
df_filtered.to_csv('sample_data/the_data_selected_timeFiltered.csv', index=False)
```
